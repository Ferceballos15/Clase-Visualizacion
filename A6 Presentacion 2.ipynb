```python
df = pd.read_excel('2019 fantasy.xlsx',
                   sheet_name = '2019',
                   engine = 'openpyxl')
df
```

```python
df.drop(['Unnamed: 0', 'G', 'GS', 'Att','Yds', 'Att.1', 'Yds.1', 'Yds.2'],
       axis = 1,
       inplace = True)
df.columns
```

```python
df.set_index('Player', 
             inplace = True)
df
```

```python
df.rename(columns = {'Tm':'Team',
                     'Pos':'Position'},
          inplace = True)
df
```
